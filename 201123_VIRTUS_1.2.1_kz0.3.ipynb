```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(100)
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib as mpl
mpl.rcParams['figure.facecolor'] = (1,1,1,1)
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
```

```python
df_cli = pd.read_csv('idep/THYM.clinical.xml.csv', index_col=0)
df_cli.index = df_cli.bcr_patient_barcode
df_cli = df_cli.dropna(subset=['primary_pathology_history_myasthenia_gravis'])
df_cli = df_cli[~df_cli.duplicated()]
df_cli.head()
```

```python
df_sample = pd.read_csv('TCGA-THYM-rawseq.csv', index_col=0)
df_sample.head()
```

```python
df_sample.loc[1]
```

```python
df_sample[::-1]
```

```python
i = 0
list_MG = []
df_VIRTUS = pd.DataFrame()
for pos,row in df_sample.iterrows():
    s = row['file_id']
    
    try:
        d = pd.read_csv('VIRTUS_1.2.1_kz0.3/{}.tsv'.format(s), sep='\t', index_col=0)
        d = d[['rate_hit']]
        d.columns = [row.cases[:12]]
        
    except FileNotFoundError as e:
        d = pd.DataFrame([[0]])
        d.columns = [row.cases[:12]]
        d.index = ['Null']
#         print(d)
    
    try:
        m = list(df_cli.loc[df_cli.bcr_patient_barcode == row['cases'][:12], 
            'primary_pathology_history_myasthenia_gravis'])[0]
    except:
        continue
        
    if not row.cases[:12] in df_VIRTUS.columns:
        if m in ['YES', 'NO']:
            if i == 0:
                df_VIRTUS = d.copy()
            else:
                df_VIRTUS = pd.merge(df_VIRTUS, d, left_index=True, right_index=True, how='outer')
            i += 1

            list_MG.append(m)

df_VIRTUS = df_VIRTUS.fillna(0)
df_VIRTUS.index = df_VIRTUS.index.str.split(',').str.get(0).str.split('_x').str.get(0)

df_VIRTUS = df_VIRTUS.drop('Null')
```

```python
col_pal = sns.color_palette()
col_colors = dict(zip(map(str, list_MG), col_pal))

dict_MG = {'NO' : 0, 'YES' : 1}
list_MG_col = [col_pal[dict_MG[x]] for x in list_MG]
```

```python
sns.clustermap(np.log10(df_VIRTUS + 10e-10), col_colors=list_MG_col, col_cluster=True,
              figsize=(10,10), cmap='viridis')
plt.savefig('plots/VIRTUS_kz0.3.pdf', bbox_inches='tight')
```

```python
df_VIRTUS_plot = df_VIRTUS.copy().T
df_VIRTUS_plot['MG'] = list_MG
df_VIRTUS_plot['vir_sum'] = df_VIRTUS_plot[df_VIRTUS.index].sum(axis=1)
df_VIRTUS_plot = df_VIRTUS_plot.sort_values(ascending=False, by=['MG', 'vir_sum'])
l_order_pat = df_VIRTUS_plot.index
l_mg = df_VIRTUS_plot['MG']
list_MG_col = [col_pal[dict_MG[x]] for x in l_mg]
```

```python
from scipy import stats

for v in df_VIRTUS.index:
    print(v)
    l_bcr_MG = df_cli.loc[df_cli.primary_pathology_history_myasthenia_gravis == 'YES', 
                          'bcr_patient_barcode']
    l_bcr_nonMG = df_cli.loc[df_cli.primary_pathology_history_myasthenia_gravis == 'NO', 
                          'bcr_patient_barcode']
    res = stats.mannwhitneyu(df_VIRTUS.loc[v, l_bcr_MG].dropna(),
    df_VIRTUS.loc[v, l_bcr_nonMG].dropna())
    print(res)
```
